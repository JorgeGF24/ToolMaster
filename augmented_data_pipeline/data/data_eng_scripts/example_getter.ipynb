```python
import pandas as pd


df = pd.read_csv('/vol/bitbucket/jg2619/augmenting_llms/augmented_data_pipeline/data/train/construction/calculator_subtype/GPTJ_huge_set/train.csv')

df.head()

```

```python
# Get a text with low relevance for each tool:
sorted_df = df.sort_values(by='loss_improvement', ascending=False)

calc_top_10 = sorted_df[sorted_df['tool_name'] == 'Calculator'][:100].sort_values(by='relevance', ascending=True)[:10].reset_index(drop=True)
wiki_top_10 = sorted_df[sorted_df['tool_name'] == 'WikiSearch'][:100].sort_values(by='relevance', ascending=True)[:10].reset_index(drop=True)
calend_top_10 = sorted_df[sorted_df['tool_name'] == 'Calendar'][:100].sort_values(by='relevance', ascending=True)[:10].reset_index(drop=True)


print_examples(calc_top_10)
print_examples(wiki_top_10)
print_examples(calend_top_10)
```

```python

def print_examples(exam_df):
    for i in range(len(exam_df)):
        print(exam_df['tool_name'][i])
        print(exam_df['API_call_response_text'][i])
        print(exam_df["loss_improvement"][i])
        print(exam_df['perplexity'][i])
        print(exam_df['relevance'][i])
        
        print()
        print(f"""{exam_df['API_call_response_text'][i].replace("<TOOL>", "8888textbf{[").replace("</TOOL>", "]}")} & {exam_df['loss_improvement'][i]} & {exam_df['perplexity'][i]} & {exam_df['relevance'][i]} & \\checkmark \\\\""".replace("8888", "\\"))
        print()
```
